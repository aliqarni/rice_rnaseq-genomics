```python
import pandas as pd

#Set gene as index and ensure counts are integers
counts = df.set_index("Gene").astype(int)

counts.shape
```

```python
libsize = counts.sum(axis = 0)
libsize.describe()
```

```python
genes_detected = (counts > 0).sum(axis=0)
genes_detected.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = "whitegrid")

```

```python
plt.figure(figsize=(8,5))
sns.histplot(libsize, bins=10, kde=True)
plt.title("Distribution of Library Sizes")
plt.xlabel("Total Counts (UMIs)")
plt.ylabel("Number of Samples")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(genes_detected, bins=10, kde=True)
plt.title("Distribution of Genes Detected")
plt.xlabel("Number of Genes Detected (>0)")
plt.ylabel("Number of Samples")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=libsize)
plt.title("Boxplot of Library Sizes")
plt.ylabel("Total Counts (UMIs)")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=genes_detected)
plt.title("Boxplot of Genes Detected")
plt.ylabel("Number of Genes Detected")
plt.show()
```

```python
meta["libsize"] = libsize.values
meta["genes_detected"] = genes_detected.values
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=meta["group"], y=meta["libsize"])
plt.title("Library Size by Group")
plt.xlabel("Group")
plt.ylabel("Total Counts")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=meta["group"], y=meta["genes_detected"])
plt.title("Genes Detected by Group")
plt.xlabel("Group")
plt.ylabel("Genes Detected")
plt.show()
```

```python
min_samples = int(0.2 * counts.shape[1]) #20% rule
min_samples
```

```python
filtered = counts [(counts > 0).sum(axis = 1)>=min_samples]
filtered.shape
```

```python
counts.shape[0] - filtered.shape[0]
```

```python
filtered_counts = filtered.copy()
```

```python
filtered.to_csv("filtered_counts.tsv", sep="\t", index=True)
meta.to_csv("metadata_for_deseq.tsv", sep="\t", index=False)
print("Files written: filtered_counts.tsv, metadata_for_deseq.tsv")
```
